```python
import matplotlib.pyplot as plt

# Extrair os valores do arquivo de texto
media_total = 53.10543235
desvio_total = 3.1
media_par = 8.7
desvio_par = 1.5

# Criar listas para os valores e rótulos
medias = [media_total, media_par]
desvios = [desvio_total, desvio_par]
labels = ['Média Total', 'Média do Par']

# Criar o gráfico de barras com barras de erro
fig, ax = plt.subplots()
ax.bar(labels, medias, yerr=desvios, align='center', alpha=0.7, ecolor='black', capsize=10)

# Adicionar rótulos e título
ax.set_xlabel('Métricas')
ax.set_ylabel('Valores')
ax.set_title('Médias com Desvios Padrão')

# Mostrar o gráfico
plt.show()
```
